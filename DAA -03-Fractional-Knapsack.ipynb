#  Knapsack using greedy strategy

````python
class item:
    def __init__(self,value,weight):
        self.value = value
        self.weight = weight
        
class solution:
    def fractionalknapsack(self,capacity,arr,n):
        
        arr.sort(key=lambda x: x.value/x.weight,reverse = True)
        
        currentw = 0
        finalvalue=0.0
        
        for i in range(n):
            if currentw + arr[i].weight <= capacity:
                currentw = currentw + arr[i].weight
                finalvalue = finalvalue+arr[i].value
            else:
                remain = capacity - currentw
                finalvalue = finalvalue + remain * arr[i].value/arr[i].weight
                break
        
        return finalvalue
    
    
if __name__ == '__main__':
    n = 3
    capacity = 50
    arr = [item(60,10),item(100,20),item(120,30)]
    obj = solution()
    ans = obj.fractionalknapsack(capacity,arr,n)
    
    print("The Maximum Value is",ans)
````
